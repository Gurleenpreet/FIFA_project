### Currency to Normal numbers 

```python
# cleaning Wage and Value columns from currency to normal numbers
def value_to_int(df_value):
    try:
        value = float(df_value[1:-1])
        suffix = df_value[-1:]

        if suffix == 'M':
            value = value * 1000000
        elif suffix == 'K':
            value = value * 1000
    except ValueError:
        value = 0
    return value

df['Value'] = df['Value'].apply(value_to_int)
df['Wage'] = df['Wage'].apply(value_to_int)
df['Release Clause'] = df['Release Clause'].apply(value_to_int)
```

### Changing into Numerical value

#### 1.Weight Conversion

```python
def weight_to_int(df_weight):
    value = df_weight[:-3]
    return value
  
df['Weight_int'] = df['Weight'].apply(weight_to_int)
df['Weight_int'] = df['Weight_int'].apply(lambda x: int(x))
```

#### 2.Height Conversion

```python
def height_to_int(df_height):
    try:
        feet = int(df_height[0])
        dlm = df_height[-2]

        if dlm == "'":
            height = round((feet * 12 + int(df_height[-1])) * 2.54, 0)
        elif dlm != "'":
            height = round((feet * 12 + int(df_height[-2:])) * 2.54, 0)
    except ValueError:
        height = 0
    return height

df['Height_int'] = df['Height'].apply(height_to_int)
```

### Graph theme will be same throught the kernel

```python
def display_graph(ax, title, xlabel, ylabel, legend):
    ax.set_xlabel(xlabel, fontsize=16)
    ax.set_ylabel(ylabel, fontsize=16)
    plt.title(title, fontsize=18, fontweight='bold')
    plt.legend(legend)
    plt.show()
```

###  Players Join as per year

```python
ax = df['Join_year'].value_counts().plot()
display_graph(ax, 'Players growth over the year', 'Years', 'Number of players', ['Players Growth'])
```

### Enrolling players as per month

```python
from datetime import datetime #importing library for strptime
df_joined = df['Joined']
df_joined = df_joined.apply(lambda x: datetime.strptime(x, '%b %d, %Y'))
df_month = df_joined.apply(lambda x: x.month)

ax = df_month.value_counts(sort = False).plot.bar(color = 'pink')
display_graph(ax, 'Enrolling players as per month', 'Month', 'Number of players', ['Player/Month'])
```
